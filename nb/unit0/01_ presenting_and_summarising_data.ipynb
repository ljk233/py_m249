# Presenting and summarising data

The first section of **M249, Unit 0** deals with various ways of presenting data visually.
We replicate the text by using various visualisations of the number of casual customers that use a bike sharing company in Washington DC in the years 2011 and 2012.
The notebook discusses the use of five different visualisations:

1. Bar charts
2. Comparative bar charts
3. Line plots
4. Scatterplots
5. Histograms

## Setup the notebook

```python
from __future__ import annotations
from pyou.stats import helpers
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load helper class
bikerental: helpers.BikeRentalDaily = helpers.BikeRentalDaily()
```

## Scenario

```python
# load views
casual_by_day: pd.DataFrame = bikerental.get_casual_by_day()
casual_by_week: pd.DataFrame = bikerental.get_casual_by_week()
casual_by_month: pd.DataFrame = bikerental.get_casual_by_month()
casual_by_atemp: pd.DataFrame = bikerental.get_casual_by_atemp()
```

## Bar charts: Casual customers by month in 2011

A bar chart is good for showing categorical data, or with numerical data with few points.
They allow for easy comparison between variables.

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.barplot(data=casual_by_month.query('year == 2011'), x="month", y="total",
            palette="deep", ci=None)
plt.xticks(rotation=70)
plt.show()
```

The bar chart shows that there were more casual customers in the Summer months when compared the Winter months, with the number of casual customers peaking in July.
This is consistent with what we'd expect, given more people have holidays in the Summer months, so are more likely to go on holidays and rent a bike.

## Comparative bar charts: Casual customers by month and year in 2011 and 2012

```python
# plot comparative bar chart
f, ax = plt.subplots(figsize=(8, 6))
sns.barplot(data=casual_by_month, x="month", y="total", hue="year")
plt.xticks(rotation=70)
plt.show()
```

## Line charts: Casual customers by week in 2011 and 2012

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(data=casual_by_week, x="week", y="total", hue="year",
             palette="deep", lw=2, style="year", markers=True, ax=ax)
ax.set(title="Add title")
plt.show()
```

## Scatterplots: Casual customers by temperature

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=casual_by_atemp, x="atemp", y="total",
                hue="year", style="year", markers=True, palette="deep")
ax.set(title="Add title")
plt.show()
```

## Histograms: Casual customers by data in 2011

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.histplot(data=casual_by_day.query('year == 2011'), x="total", ax=ax)
ax.set(title="Casual customers by day in 2011")
plt.show()
```
